## Предобработка данных

### Переименование столбцов

```python
df.columns = ['total_images', 'last_price', 'total_area', 'first_day_exposition',
       'rooms', 'ceiling_height', 'floors_total', 'living_area', 'floor',
       'is_apartment', 'studio', 'open_plan', 'kitchen_area', 'balcony',
       'locality_name', 'airports_nearest', 'city_centers_nearest',
       'parks_around3000', 'parks_nearest', 'ponds_around3000',
       'ponds_nearest', 'days_exposition']
```

### Обработка пропусков

####  `is_apartament`

Больше всего пропусков (88% от массива) в графе с указанием является ли недвижимость апартаментами. Посмотрим распределение  по этому параметру. 

```python
df['is_apartment'].value_counts(dropna=False) / len(df)
```

Всего 0.2% недвижимости являются апартаментами. И скорее всего в нашем случае 88% пропусков апартаментами не являются, а пользователи при заполнении данных просто пропускали этот пункт, не считая необходимым указывать тот факт, что недвижимость не является апартаментами.

Заполним пропуски в данных значением False.

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
df['is_apartment'].value_counts(dropna=False)
```

#### `ponds_around3000`, `parks_around3000`

Рассмотрим причины пропусков в данных о количествах водоемов и парков в радиусе 3 км. Скорее всего эти пропуски связаны с отсутствием таковых. Посмотрим распределение пропусков по населенным пунктам. И тогда, согласно нашему предположению о причине появления пропусков, в этом списке не будет Санкт-Петербурга, как города с хорошо развитой инфраструктурой, а будут населенные пункты Ленинградской области.

```python
df.loc[(df['ponds_around3000'].isna()) & (df['parks_around3000'].isna())]['locality_name'].value_counts().head(20)
```

Так оно и получилось. Предположение о том, что пропуски в данных означают отсутствие в радиусе 3 км парков и водоемов близко к правде. Заменим пропуски на `0`

```python
df[['ponds_around3000', 'parks_around3000']] = df[['ponds_around3000', 'parks_around3000']].fillna(0)
df[['ponds_around3000', 'parks_around3000']].isna().sum()
```

Пропусков больше нет.

#### `parks_nearest`, `ponds_nearest`, `airports_nearest`, `city_centers_nearest`

Логично предположить, что пропуски в данных о расстояниях до ближайших парка, водоема, аэропорта и центра города, связаны с  тем, что это расстояния достаточно велики. Посмотрим на количество ближайших парков и водоемов в данных где есть пропуски в расстоянии до них. Согласно нашему предположению там должны быть 0. 

```python
df[df['parks_nearest'].isna()]['parks_around3000'].value_counts()
```

```python
df[df['ponds_nearest'].isna()]['ponds_around3000'].value_counts()
```

Действительно так. Где пропущены данные о расстоянии до ближайших парка и водоема, их попросту нет в пределах 3 км. Значит это расстояние больше 3 км.

Что касается расстояния до ближайшего аэропорта и центра города, то тут можно посмотреть распределение данных с пропусками в расстоянии по населенным пунктам. И согласно нашему предположению, что пропуски связаны с большой величиной этого расстояния, в списке с пропусками в данных должны отобразится удаленные от Санкт-Петербурга населенные пункты.

```python
df[(df['airports_nearest'].isna()) & (df['city_centers_nearest'].isna())]['locality_name'].value_counts().head(20)
```

Все действительно так как мы предполагали. Эти пропуски мы заполнять не будем, но мы будем держать в голове, что пропуски в данных о расстояниях до ближайших парка, водоема, аэропорта и центра города, означают, что расстояние достаточно велико.

#### `balcony`

В данных о количестве балконов также довольно много пропусков (~ 50%). Скорее всего люди при заполнении данных в случае отсутствия балкона просто пропускали данный пункт, оставляя его незаполненным.
Заполним пропуски в данных на `0`

```python
df['balcony'] = df['balcony'].fillna(0)
df['balcony'].value_counts(dropna=False)
```

Пропусков больше нет и мы еще заодно узнали, что у кого-то целых пять балконов)

#### `ceiling_height`

В данных о высоте потолков также присутствуют пропуски в достаточно большом количестве (~ 40%). В данном случае можно заменить пропуски на медиану (в данном случае лучше использовать медиану, чем среднее, так как присутствуют аномальные данные 100м).

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
df['ceiling_height'].isna().sum()
```

#### `days_exposition`, `locality_name`, `floors_total`

Пропуски в данных о количестве дней размещения (13% от всех данных) обьявления могут быть связаны с тем, что на момент выгрузки данных обьявление еще не было снято с публикации. Не зная точную природу возникновения пропусков, мы не будем их заполнять в данном столбце.

Пропуски в данных о населенном пункте и количествве этажей в общей сложности составляют долю меньше 1% и очевидного способа их заполнения также нет.

```python
df[df['floors_total'].isna()]
```

#### `kitchen_area`, `living_area`

В данных о жилой площади и площади кухни имеются пропуски в количестве около 10% по каждому пункту.

Для заполнения этих пропусков можно воспользоваться данными об общей площади недвижимости, предварительно проверив корреляцию этих показателей и в случае наличии посчитать коэффициент отношения данных площадей к общей площади. Только нужно проверить один момент: в данных по площадям кухни могут быть умышленные пропуски, связанные с тем, что недвижимость является студией и соответственно не имеет отдельной кухни как таковой. Проверим это предположение.


```python
df[df['studio'] == True]['kitchen_area'].value_counts(dropna=False)
```

Действительно наше предположение верно - в случае если недвижимость является студией, в графе с площадью кухни преднамеренно стоят пропуски, так как кухни нет, и соответственно пропуски в этом случае не нужно заполнять.

Посчитаем корреляции для данных о жилой площади и площади кухни.

```python
df.corr()[['living_area', 'kitchen_area']].sort_values('living_area', ascending=False)
```

```python
df.plot(kind='scatter', y='living_area', x='total_area')
```

```python
df.plot(kind='scatter', y='kitchen_area', x='total_area')
```

```python
df[df['is_apartment'] == True]
```

Действительно жилая площадь сильно коррелирует с общей площадью, площадь кухни в меньшей степени, но все таки связь довольно сильна. Тогда для заполнения пропусков можно рассчитать отношения известных нам площадей и применить их к пропущенным значениям(за исключением площади кухни для студий)

```python
coef_kitchen_from_total = (df['kitchen_area'] / df['total_area']).median()
coef_living_from_total = (df['living_area'] / df['total_area']).median()
coef_kitchen_from_total, coef_living_from_total
```

```python
df['living_area'] = df['living_area'].mask(df['living_area'].isna(), df['total_area'] * coef_living_from_total)
```

```python
df['kitchen_area'] = df['kitchen_area'].mask((df['kitchen_area'].isna()) & (df['studio'] == False), df['total_area'] * coef_kitchen_from_total)
```

```python
df['living_area'].isna().sum()
```

```python
df['kitchen_area'].isna().sum()
```

```python
len(df[df['studio'] == True])
```

```python
len(df[(df['kitchen_area'].isna()) & (df['studio'] == False)])
```

Все пропуски в графе жилой площади заполнены, а в графе с площадью кухни остались пропуски только для студий.

Посмотрим на уникальные значения в данных о населенных пунктах

```python
df['locality_name'].sort_values().unique()
```

```python
len(df['locality_name'].unique())
```

Всего среди наименований населенных пунктов сейчас 365 уникальных значений, но среди них присутствуют дубликаты.

Две причины возникновения дубликатов:
* использование буквы `ё`
* разные варианты написания `поселок` (поселко городского типа и т.д.)

```python
# создадим список с возможными дубликатами "поселок"
locality_duplicates = ['городской поселок', 'коттеджный поселок', 'поселок городского типа', 'поселок при железнодорожной станции']   

# заменим 'ё' на 'е'
df['locality_name'] = df['locality_name'].str.replace('ё', 'е')

# заменим все возможные дубликаты на значение "поселок"
for locality_duplicate in locality_duplicates:
    df['locality_name'] = df['locality_name'].str.replace(locality_duplicate, 'поселок')

len(df['locality_name'].unique())    
```

Все скрытые дубликаты в наименованиях населенных пунктов убраны (их было порядка 12%)

### Добавление данных о расположении этажа квартиры

Добавим данные является этаж первым, последним или другим (ни первым, ни последним)

```python
def floor_disposition(row):
    if row['floor'] == 1:
        return 'first'
    if row['floor'] == row['floors_total']:
        return 'last'
    else:
        return 'other'
```

```python
df['floor_position'] = df.apply(floor_disposition, axis=1)
df['floor_position'].value_counts(dropna=False)
```

```python
# добавим графу с расположенеим этажа квартиры
df['floor_position'] = (df['floor'].mask(df['floor'] == 1, 'first') # первый этаж
                        .mask(df['floor'] == df['floors_total'], 'last') # последний этаж
                        .mask((df['floor'] != df['floors_total']) & (df['floor'] != 1), 'other')) # ни первый, ни последний этаж
```

Данные о расположении этажа квартиры добавлены.

### Разделение объявлений по локации

Мы уже выяснили, что квартиры в Петербурге отличаются по стоимости от остальных населенных пунктов. Теперь проанализируем квартиры непосредственно в Санкт_петербурге.

Глядя на графики можно уточниить зону в пределах 7 км с наиболее высокими ценами на жилье. Это скорее всего и будет центром города. 

Исходя из этого можно разделить наши данные на три группы по локации:
* центр Санкт-петербурга
* Санкт-Петербург
* пригород Санкт-Петербурга

Добавим к нашим данным метки о распложении объекта: центр, город, пригород.

```python
def distribution_by_location(row):
    if row['city_centers_nearest'] <= 7000:
        return 'center'
    elif row['locality_name'] == 'Санкт-Петербург':
        return 'city'
    else:
        return 'suburb'
```

```python
df_cleaned['locality_name'].unique()
```

```python
df_cleaned['locality_group'] = df_cleaned.apply(distribution_by_location, axis=1)
```

```python
df_cleaned['locality_group'].value_counts() / len(df_cleaned)
```

Распределение по количеству объявлений получилось следующим:
* 12% - в центре города
* 55% - в Санкт-Петербурге (за исключением центра)
* 33% - в пригороде
