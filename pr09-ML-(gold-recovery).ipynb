### Схема технологического процесса

Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.
1. Флотация  
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. Очистка  
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Описание данных

**Технологический процесс**
* Rougher feed — исходное сырье
* Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
** Xanthate — ксантогенат (промотер, или активатор флотации);
** Sulphate — сульфат (на данном производстве сульфид натрия);
** Depressant — депрессант (силикат натрия).
* Rougher process (англ. «грубый процесс») — флотация
* Rougher tails — отвальные хвосты
* Float banks — флотационная установка
* Cleaner process — очистка
* Rougher Au — черновой концентрат золота
* Final Au — финальный концентрат золота

**Параметры этапов**
* air amount — объём воздуха
* fluid levels — уровень жидкости
* feed size — размер гранул сырья
* feed rate — скорость подачи

### Наименование признаков

Наименование признаков должно быть такое:  
`[этап].[тип_параметра].[название_параметра]`  
Пример:  
`rougher.input.feed_ag`

Возможные значения для блока `[этап]`:
* rougher — флотация
* primary_cleaner — первичная очистка
* secondary_cleaner — вторичная очистка
* final — финальные характеристики

Возможные значения для блока `[тип_параметра]`:
* input — параметры сырья
* output — параметры продукта
* state — параметры, характеризующие текущее состояние этапа
* calculation — расчётные характеристики

### Проверка расчета метрики recovery

Рассчитаем по известным нам параметрам метрику эффективности на стадии флотации для обучающей выборки

```python
train["rougher.output.recovery_manual"] = (
    (
        train["rougher.output.concentrate_au"]
        * (train["rougher.input.feed_au"] - train["rougher.output.tail_au"])
    )
    / (
        train["rougher.input.feed_au"]
        * (train["rougher.output.concentrate_au"] - train["rougher.output.tail_au"])
    )
    * 100
)
```

Сравним подсчитанный вручную показатель эффективности флотации с данными из выборки. Для сравнения используем метрику MAE (средняя абсолютная ошибка) (формулу напишем вручную)

```python
abs(train["rougher.output.recovery"] - train["rougher.output.recovery_manual"]).mean()
```

И все таки подверимся метрикой из sklern

```python
mean_absolute_error(
    train["rougher.output.recovery"], train["rougher.output.recovery_manual"]
)
```

```python
mean_absolute_percentage_error(
    train["rougher.output.recovery"], train["rougher.output.recovery_manual"]
)
```

Средняя абсолютная ошибка ничтожно мала как в абсолютном, так и в относительном виде.

## Анализ данных

### Анализ изменения концентрации металлов (Au, Ag, Pb) на различных этапах очистки

Посмотрим на изменение концентрации металлов в продукте после различных стадий обогащения

```python
# словарь с названиями стадий обогащения
stage_name_dict = {
    "rougher.input": "исходное сырье",
    "rougher.output": "после флотации",
    "primary_cleaner.input": "перед первой очисткой",
    "primary_cleaner.output": "после первой очистки",
    "secondary_cleaner.output": "после второй очистки",
    "final.output": "финал",
}
```

```python
metall = ["au", "ag", "pb"]
stage_to_analise = [
    "final.output",
    "primary_cleaner.output",
    "rougher.output",
    "rougher.input",
]
```

```python
fig, ax = plt.subplots(3, 1, figsize=(10, 20))

for i in range(3):
    for y in range(len(stage_to_analise) - 1):
        ax[i].hist(
            full[f"{stage_to_analise[y]}.concentrate_{metall[i]}"],
            label=stage_name_dict[stage_to_analise[y]],
            alpha=1 - (y * 0.2),
        )
    ax[i].hist(
        full[f"{stage_to_analise[-1]}.feed_{metall[i]}"],
        label=stage_name_dict[stage_to_analise[-1]],
        alpha=1 - (3 * 0.2),
    )
    ax[i].set_title(
        f"Распределение концентраций {metall[i].upper()} после стадий обогащения"
    )

    ax[i].legend(title="cтадия")
    ax[i].grid()
plt.show()
```

На гистограммах распределений концентрации золота четко видно, что с каждой следующей стадией обогащения концентрация металла увеличивается, что вполне логично, ведь как раз цель обогащения и заключается в том, чтобы сконцентрировать максимально возможное количество золота в минимально возможном объеме продукта (увеличение концентрации относительно исходного сырья). Тот факт, что концентрация золота от стадии к стадии увеличивается говорит нам о том, что в глобальном понимании процесс обогащения настроен правильно, в дальнейшем нужно только оценить его качество.

На гистограммах распределения серебра уже все не так однозначно. Видно, что после флотации концентрация серебра увеличивается, но на стадиях очистки серебро уже уходит в хвостовые продукты и концентрация металла падает. Если бы целью процесса обогащения кроме золота было бы серебро (случай когда вместе с золотом на фабрике извлекают попутно серебро как попутный ценный компонент), то мы бы в этом случае сказали бы о недопустимых потерях металл на стадии очистки.

Для свинца ситуация похожа на ситуацию для серебра, с той лишь разницей, что на стадиях очистки концентрация свинца не падает.

По данным гистограммам четко видно, что процесс обогащения настроен на извлечение именно золота из исходного сырья (а если говорить точнее на увеличение его концентрации в финальном продукте)

### Сравнение распределения размеров гранул сырья на обучающей и тестовой выборках

```python
fig, ax = plt.subplots(2, 1, figsize=(12, 10))
stage_feed = ["rougher.input", "primary_cleaner.input"]
for i in range(2):
    ax[i].hist(
        train[f"{stage_feed[i]}.feed_size"], label="train", bins=20, density=True
    )
    ax[i].hist(
        test[f"{stage_feed[i]}.feed_size"],
        alpha=0.8,
        label="test",
        bins=20,
        density=True,
    )

    ax[i].set_title(f"Распределение размеров гранул\n {stage_name_dict[stage_feed[i]]}")

    ax[i].legend(title="выборка")
    ax[i].grid()
plt.show()
```

## Модель

### Метрика качества

Для оценки качества нашей модели будем использовать метрику sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»). Напишем функцию для ее расчета.

```python
def smape(y_true, y_pred):
    return (
        np.mean(abs(y_true - y_pred) / ((abs(y_true) + abs(y_pred)) / 2))
        * 100
    )
```

```python
def smape_total(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```

Создадим кастомный оценщик куда присвоим нашу метрику sMAPE

```python
smape_score = make_scorer(smape, greater_is_better=False)
```

```python
def print_score(smape_rougher, smape_final):
    print(f"sMAPE rougher_recovery - {smape_rougher:.2f}")
    print(f"sMAPE final_recovery - {smape_final:.2f}")
    print(f"sMAPE total - {smape_total(smape_rougher, smape_final):.2f}")
```
