## Описание проекта

В нашем распоряжении исторические данные о продажах игр по регионам, оценки пользователей и экспертов, жанры и платформы.

### Описание данных

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)

- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

### Анализ пропусков в годах релиза игр

Как мы уже выяснили ранее в 1.6% случаях у нас пропуски в годе релиза игры. Причем для одной и той же игры, на одной платформе могжет быть указан год релиза, а на другой платформе - нет. В этом случае логично предположить, что обычно одна и та же игра выпускается на платформах в одно и то же время (+/- год). Если наше предположение верно, то часть пропусков можно заполнить по известным годам релиза этой же игры, но на других платформах.

```python
# проверим наше предположение о том, что на разных платформах года релиза одной и тойже игры не сильно различаются
# сгруппируем игры по названию и посчитаем для каждой кол-во релизов и минимальные и максимальный год релиза
t = (df.groupby('name')['year_of_release'].agg(['count', 'min', 'max'])
     .query('count > 1')) # отфильтруем игры с кол-ом релизов больше одного для представительности анализа
t['diff'] = t['max'] - t['min'] # вычислим разницу между минимальным и максимальным годом релиза (разброс между релизами)
t['diff'].value_counts(normalize=True) # посмотрим на разброс между релизами
```

Наше предположение верно: почти в 90% случаях между релизами одной и той же игры на разных платформах проходит не более 1 года. Соответственно с небольшой вероятностью совершить ошибку, часть пропусков можно заполнить по известным годам релиза этой же игры, но на других платформах.

```python
# создадим словарь ~игра - медианное значение года релиза~ (для известных значений года релиза)
years_dict = df[df['year_of_release'].notna()][['name','year_of_release']].groupby('name').median().astype('int').reset_index()
years_dict.head()
```

```python
# добавим к нашему датафрейму столбец со значениями из словаря для заполнения NaN
df = df.merge(years_dict, on='name', how='left', suffixes=('', '_for_nan'))
```

```python
print (f"Количество пропусков в годах релиза до заполнения - {len(df[df['year_of_release'].isna()])}")
```

```python
# заполним пропуски по уже подготовленному для этого столбцу
df['year_of_release'] = df['year_of_release'].fillna(df['year_of_release_for_nan'])
```

```python
print (f"Количество пропусков в годах релиза после заполнения - {len(df[df['year_of_release'].isna()])}")
```

Больше половины пропусков удалось заполнить таким методом

Также встречаются случаи, когда в названии игры указан год (например `FIFA 2011`). Это означает, что игра вышла либо в этот год, но чаще всего годом ранее (частый случай для спортивных игр). Часть пропусков в годах еще можно заполнить, взяв год из назания игры (он всегда в конце названия).

```python
# выведем на экран игры с пропуском в графе дата релиза в конце названия которых последние 4 символа цифры 
# это и будет год релиза (+/- год)
df[(df['year_of_release'].isna()) & (df['name'].str.slice(-4).str.isdigit())]['name']
```

```python
# напишем для этих игр функцию заполнения года из названия игры
def year_from_name(row):
    if (np.isnan(row['year_of_release']) and row['name'][len(row['name'])-4:].isdigit()): # наше условие
        return ((int(row['name'][len(row['name'])-4:]) - 1)) # заполняем год последними четырьмя цифрами из названия - год 
    return row['year_of_release']
```

```python
df['year_of_release'] = df.apply(year_from_name, axis=1)
```

```python
# удалим ненужные столбцы
df = df.drop('year_of_release_for_nan', axis=1)
```

### Анализ пропусков в оценках критиков и пользователей

Как мы видим, что в основном оценки критиков и пользователей отсутствуют для одних и тех же игрЖ:

- в 93% (77.5 + 15.5) случаях если нет оценки критиков, то нет оценки и пользователей

Почти в 15% случаях в графе оценки пользователей стоит значение `tbd` (*to be determind*), что означает: "рейтинг калибруется - пока точно не  определен" (скорее всего малое количество оценок для определения среднего, в связи с тем что игра недвано поступила в продажу). Можно быстро посмотреть на среднее продаж этих игр и сравнить с остальными.

Да действительно средние продажи игр с аббривиатурой tdb в три раза меньше, значит они находятся на стадии калибровки оценки.

Пропуски же в оценках скорее всего связаны, что система оценок ввелась позже, чем начало периода, за который представлены данные по продажам и для "древних" игр оценок просто нет.

```python
# добавим временную маску на отсутсвтие оценки
df['user_score_is_nan'] = df['user_score'].isna()
```

Предположение верно: у подавляющего большинства игр до 2000 года отсутствуют оценки. А игр после 2000 года отсутствие оценки связано скорее всего с их не очень высокой популярностью.

Да действительно продажи игр с оценкой в среднем в два раза больше, чем игр без оценки. Наше предположение о пропусках отчасти верно, но в любом случае заполнить их объективно не видится возможным. Единственное, что необходимо сделать - это заменить строковое значение `tbd` на `NaN`(по сути случаи для нас идентичны: в обоих случаях нет оценки пользователей) для того, чтобы в последующем можно было работать с оценками как с числами.

Соответственно есть зависимость - "чем выше оценка, тем выше продажи", причем эта зависимость более ярко выражена для оценки критиков (чем выше оценка критиков, тем успешнее игра) и она характерна больше для `PS3` и `X360`, чем для `Wii`. **Успешность игры в объеме продаж зависит от оценки критиков**

Посмотрим еще на таблицу корреляций по каждой платформе для показателей общих продаж и оценок критиков и пользователей.

```python
 # таблица корреляций  общих продаж по платформам с оценками критиков и пользователей
(df_actual.groupby('platform')[['total_sales', 'critic_score']].corr()['total_sales'].reset_index() 
 .query("level_1 == 'critic_score'")                                                               # с оценкой критиков
 .merge(df_actual.groupby('platform')[['total_sales', 'user_score']].corr()['total_sales'].reset_index() # соединяем
 .query("level_1 == 'user_score'"),                                                           # с оценкой пользователей
        on='platform', suffixes=('_to_critic_score', '_to_user_score'))                       # добавляем суффиксы
 .drop(['level_1_to_critic_score', 'level_1_to_user_score'], axis=1)                       # и удаляем лишние столбцы
 .sort_values('total_sales_to_critic_score', ascending=False))                         
```

Исходя из этой логики связь между оценками критиков и пользователей должная быть достаточно сильной, ведь в идеале они должны сильно коррелировать между собой - если игра хороша, то она хороша и для тех, и для других.

Для всех гипотез критический уровень статистической значимости примем 0.05.

### Пользовательские оценки Xone и PC

Существует предположение, что средние пользовательские рейтинги платформ **XOne** и **PC** одинаковые. Проверим это.

Сформулируем нулевую гипотезу: **Н0** - средняя оценка пользователей игр платформы **XOne** **равна** средней оценке пользователей игр платформы **PC**  
Тогда альтернативная гипотеза: **H1** -  средняя оценка пользователей игр платформы **XOne** **не равна** средней оценке пользователей игр платформы **PC**

```python
# напишем функцию для проверки гипотез
def hypothesis_ttesting(parameter, # по какому параметру смотрим
                        x, # значение 'x' параметра 
                        y, # значение 'x' параметра
                        target, # цель, что сравниваем
                        data=df_actual, # массив данных
                        alpha=alpha):
    # проведем t-тест для наших выборок при этом указав, что дисперсии для выборок не равны
    _, pvalue  = (st.ttest_ind((data[data[parameter].isin(x)][target].dropna()), 
                            (data[data[parameter].isin(y)][target].dropna()), 
                            equal_var = False))
    # сравним получившиеся p-value c пороговым значением alpha
    print(f'p-value = {pvalue}')
    if pvalue < alpha:
        print ('Отвергаем нулевую гипотезу')
    else:
        print ('Не отвергаем нулевую гипотезу')
    print(f'Среднее значение {target} для {x} - {data[data[parameter].isin(x)][target].mean():.2f}')
    print(f'Среднее значение {target} для {y} - {data[data[parameter].isin(y)][target].mean():.2f}')
```

```python
hypothesis_ttesting(parameter='platform', 
                    x=['XOne'],  
                    y=['PC'], 
                    target='user_score')
```

P-value больше заданного критического уровня значимости, что означает "При условии верной нулевой гипотезы о равенстве средних пользовательских оценок игр для XOne и PC, вероятность получить такие значения выборок по оценкам как у нас менее 16%, следовательно мы не отвергаем нулевую гипотезу. Из этого следует, что **средняя оценка пользователей игр для XOne равна средней оценке PC**

Теперь приступим к интересующим нас гипотезам. В ходе анализа исследовательского анализа данных мы сделали некоторые выводы и предположения, которые бы хорошо проверить статистикой.
